## 2. Wieviele Registrierungen gab es zu Beginn von Corona im Jahr 2020? Haben sich seit 2020 im Mittel mehr oder weniger neue Nutzer registriert als vor der Pandemie?

```python
df.columns
```

```python
print(len(df[df['Year Patron Registered'] == 2020]))
```

```python
df['Year Patron Registered'] = pd.to_numeric(df['Year Patron Registered'], errors='coerce')
```

```python
df_by_year = df['Year Patron Registered'].value_counts()
df_by_year
```

```python
registered_per_year=df.loc[
         (df['Year Patron Registered'] == 2003)|
    (df['Year Patron Registered'] == 2004)|
    (df['Year Patron Registered'] == 2005) | 
    (df['Year Patron Registered'] == 2006)|
    (df['Year Patron Registered'] == 2008)|
    (df['Year Patron Registered'] == 2009)|
     (df['Year Patron Registered'] == 2010)|
    (df['Year Patron Registered'] == 2012) | 
    (df['Year Patron Registered'] == 2013)|
    (df['Year Patron Registered'] == 2014)|
    (df['Year Patron Registered'] == 2015) | 
    (df['Year Patron Registered'] == 2016)|
    (df['Year Patron Registered'] == 2017)|
    (df['Year Patron Registered'] == 2018)|
     (df['Year Patron Registered'] == 2019)|
     (df['Year Patron Registered'] == 2020)|
    (df['Year Patron Registered'] == 2021)|
    (df['Year Patron Registered'] == 2022)  ] 

#Hilfsvariable für den Plot, damit nur die Merkmale 'email' und 'print' angezeigt werden

```

```python

sns.catplot(x='Year Patron Registered', kind='count', data=registered_per_year,  aspect=3,)
```

```python
registered_before_2020 = df[(df['Year Patron Registered'] <= 2019)]
registered_before_2020
```

```python
registered_before_2020_mean = registered_before_2020['Year Patron Registered'].value_counts().mean()
registered_before_2020_mean
```

```python
registered_before_2020_median = registered_before_2020['Year Patron Registered'].value_counts().median()
registered_before_2020_median
```

```python
registered_after_2019 =  df.loc[(df['Year Patron Registered'] >= 2020)]
registered_after_2019
```

```python
df_registered_after_2019_mean = registered_after_2019['Year Patron Registered'].value_counts().mean()
df_registered_after_2019_mean
```

```python
registered_after_2019_median = registered_after_2019['Year Patron Registered'].value_counts().median()
registered_after_2019_median
```

```python
data_mean_registration = { 'mean_before_2020': [registered_before_2020['Year Patron Registered'].
                                                value_counts().mean()],
                          'mean_after_2019': [registered_after_2019['Year Patron Registered'].value_counts().mean()]
                       }
df_mean_registration = pd.DataFrame(data=data_mean_registration)

df_mean_registration
```

```python
data_mean_registration = { 'mean_before_2020': [registered_before_2020['Year Patron Registered'].
                                                value_counts().mean()],                          
                          'mean_after_2019': [registered_after_2019['Year Patron Registered'].
                                              value_counts().mean()],
                          'median_after_2019': [registered_after_2019['Year Patron Registered'].
                                              value_counts().median()],
                          'median_before_2020': [registered_before_2020['Year Patron Registered'].
                                                value_counts().median()]
                       }
df_mean_registration = pd.DataFrame(data=data_mean_registration)

df_mean_registration
```

```python
import matplotlib.pyplot as plt 


df_mean_registration.plot(y=["mean_before_2020", "mean_after_2019", 
                             "median_before_2020",
                                "median_after_2019", 
                                ],
                                kind="bar",figsize=(9,8))

plt.show()
```

### Im Jahr 2020 haben sich 32.999 Kunden registriert. - Seit 2020 ist die Anzahl der Registrierungen sowohl im Mittelwert als auch im Median im Vergleich zu vorher gestiegen.
