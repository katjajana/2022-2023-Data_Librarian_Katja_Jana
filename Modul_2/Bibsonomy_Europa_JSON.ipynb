```python
import pandas as pd
import json
```

```python
# einlesen von Daten aus Bibsonomy
import requests
url = "https://www.bibsonomy.org/json/search/europa%20?items=1000"
result = requests.get(url) # result ist ein requests.models.Response Objekt
data = result.json() # das Response-Objekt bietet diese nützliche Methode an

```

```python
data.keys()
```

```python
type(data["items"])
```

```python
data["items"][5]
```

```python
df_bibsonomy_Europa = pd.DataFrame(data["items"])
```

```python
df_bibsonomy_Europa
```

```python

df_bibsonomy_Europa_publications = df_bibsonomy_Europa.loc[df_bibsonomy_Europa['type'] == 'Publication']
```

```python
df_bibsonomy_Europa_publications
```

```python
df_bibsonomy_Europa_publications.columns
```

```python
df_bibsonomy_Europa_publications['label'].isnull().sum()
```

```python
df_bibsonomy_Europa_selection = df_bibsonomy_Europa_publications[
    ["type", "id", "tags", "label", "user", "description", "date", "authors", "publisher", "isbn"]] 
```

```python
df_bibsonomy_Europa_selection
```

```python
df_bibsonomy_Europa_selection = df_bibsonomy_Europa_selection.rename(columns={"label": "title"})
df_bibsonomy_Europa_selection
```
