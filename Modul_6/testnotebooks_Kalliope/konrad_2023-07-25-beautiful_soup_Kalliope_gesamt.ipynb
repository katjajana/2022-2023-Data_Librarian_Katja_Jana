```python
import os

from bs4 import BeautifulSoup

import pandas as pd

import re
```

```python
path = '/home_ext/PK/b-kj102/Dokumente/ZBIW_datalib_Projekt_Kalliope/testset_ead'
# path ="."

titles = []
authors_lists = []
list_ead = []
isbns = []


for filename in os.listdir(path):
    if filename.endswith(".xml"):
        fullpath = os.path.join(path, filename)
        print(fullpath)
        soup = BeautifulSoup(open(fullpath), 'xml')
        list_ead.append(soup)
        print("tite:", soup.title)
        try:
            titles.append(soup.title.string)
        except AttributeError:
            titles.append("no title")
        # print(soup.find_all('repository'))
        cur_authors = []
        for person in soup.find_all('persname'):
            print("-", person.string)
            cur_authors.append(person.string)
            try:
                print("  -", person["authfilenumber"])
                print("----")
            except KeyError:
                print("  - no authfilenumber")
            authors_lists.append(" ,".join)   
        print(soup.ead["xmlns"])
        isbns.append(soup.ead["xmlns"])
        print("=======================")
```
